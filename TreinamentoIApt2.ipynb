#Correlation measures the relationship between two variables.
* Values Between -1 and 1
 * 1 = there is a perfect linear relationship between the variables
 * 0 = there is no linear relationship between the variables
 * -1 = there is a perfect negative linear relationship between the variables

**There is an important difference between correlation and causality:**
* Correlation is a number that measures how closely the data are related
* Causality is the conclusion that x causes y.

**Correlation Does Not Imply Causality**

* During the summer, the sale of ice cream at a beach increases
* Simultaneously, drowning accidents also increase as well
* Does this mean that increase of ice cream sale is a direct cause of increased drowning accidents?

```python
import pycountry_convert as pc

def get_continent(country_name):
    try:
        # Ajustes manuais para nomes não reconhecidos
        if country_name == 'South Korea': country_name = 'Korea, Republic of'
        if country_name == 'North Korea': country_name = "Korea, Democratic People's Republic of"
        if country_name == 'Taiwan': country_name = 'Taiwan, Province of China'
        if country_name == 'Russia': country_name = 'Russian Federation'
        if country_name == 'Iran': country_name = 'Iran, Islamic Republic of'
        if country_name == 'Venezuela': country_name = 'Venezuela, Bolivarian Republic of'
        if country_name == 'Moldova': country_name = 'Moldova, Republic of'
        if country_name == 'Syria': country_name = 'Syrian Arab Republic'

        country_code = pc.country_name_to_country_alpha2(country_name)
        continent_code = pc.country_alpha2_to_continent_code(country_code)
        continents = {
            'AF': 'Africa',
            'AS': 'Asia',
            'EU': 'Europe',
            'NA': 'North America',
            'SA': 'South America',
            'OC': 'Oceania'
        }
        return continents.get(continent_code, 'Other')
    except:
        return 'Other'

df['Continent'] = df['Country'].apply(get_continent)
```

```python
qi_continente = df.groupby('Continent')['IQ_2024'].mean().sort_values(ascending=False)
print(qi_continente)
```

```python
import matplotlib.pyplot as plt

df_sorted = df.sort_values(by='IQ_2024', ascending=False)

plt.figure(figsize=(18, 8))
plt.bar(df_sorted['Country'], df_sorted['IQ_2024'], color='cornflowerblue')
plt.xticks(rotation=90)
plt.ylabel('QI Médio (2024)')
plt.title('QI médio por país - 2024')
plt.tight_layout()
plt.show()
```

```python
qi_continente = df.groupby('Continent')['IQ_2024'].mean().sort_values(ascending=False)

#fazer o garfico
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 6))
qi_continente.plot(kind='bar', color='mediumseagreen')
plt.ylabel('QI Médio (2024)')
plt.title('QI médio por continente - 2024')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
df_BLI = pd.read_csv('/content/BLI.csv', sep=',', header=0)
df_GDP = pd.read_csv('/content/GDP.csv', sep=',', header=0)


print(df_GDP.columns.tolist())
print(df_BLI.columns.tolist())


df_GDP['Country '] = df_GDP['Country '].str.strip()
df_BLI['Country'] = df_BLI['Country'].str.strip()


print(df_BLI.columns.tolist())

# Removendo espaços extras de todas as colunas do df_GDP
df_GDP.columns = df_GDP.columns.str.strip()


print(df_GDP.columns.tolist())

# fazer o join entre os 3
df_joined = pd.merge(df[['Country', 'IQ_2024']], df_BLI[['Country', 'OBS_VALUE']], on='Country', how='inner')

# Selecionar o GDP de 2019
df_GDP_2019 = df_GDP[['Country', '2019']]
df_GDP_2019 = df_GDP_2019.rename(columns={'2019': 'GDP'})

# Realizrar o JOIN com o df_GDP_2019
df_joined = pd.merge(df_joined, df_GDP_2019[['Country', 'GDP']], on='Country', how='inner')


df_joined = df_joined.rename(columns={'OBS_VALUE': 'BLI'})


print(df_joined.head())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
# Verificar a correlação entre QI e BLI
correlacao = df_joined[['IQ_2024', 'BLI', 'GDP']].corr()
print("Matriz de correlação entre QI, BLI e GDP:\n", correlacao)


sns.scatterplot(data=df_joined, x='BLI', y='IQ_2024')
plt.title('Relação entre BLI e QI')
plt.show()
#N tem correção entre os 3, apenas os iq, e bli(ou a correação do gpd entre eles seja mt baixa.)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Criar um gráfico de dispersão para visualizar a correlação entre QI e BLI
plt.figure(figsize=(8, 6))
sns.scatterplot(data=df_joined, x='BLI', y='IQ_2024')
plt.title('Correlação entre QI e BLI por país')
plt.xlabel('BLI')
plt.ylabel('QI Médio (2024)')
plt.tight_layout()
plt.show()
```
