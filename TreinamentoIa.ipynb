```python
#Abrindo o arquivo

import pandas as pd

df = pd.read_csv("train.csv")
df
```

```python
df.iloc[0] # primeiro digito
```

```python
print(df.columns.tolist())
```

```python
# excluindo o rótulo
x = df.drop("label", axis=1)
y = df['label']
```

```python
x # features
```

```python
y # label
```

```python
# exibindo no formato de imagem um digito

image0 = x.iloc[0].to_list()

for i in range(28):
    for j in range(28):
        pixel = image0[i*28 + j]
        if pixel < 10: space = '   '
        elif pixel < 100: space = '  '
        else: space = ' '
        pixel = image0[i*28 + j]
        print(pixel, end=space)
    print()

print('Classe:',df.iloc[0]['label'])
```

```python
from matplotlib import pyplot as plt
import numpy as np
first_image = np.array(image0, dtype='float')
pixels = first_image.reshape((28, 28))
plt.imshow(pixels, cmap='gray')
plt.show()
```

```python
#dividindo entre conjunto de treino e testes

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.20, random_state=42)
```

```python
# normalizando os dados

from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
```

```python
X_train_scaled
```

```python
X_train_scaled[0]
```

```python
# treinando os modelos

from sklearn.tree import DecisionTreeClassifier
clf = DecisionTreeClassifier().fit(X_train_scaled, y_train)

# testar com diferentes parametros, modelos diferentes (principalmente deep learning com redes CNNs),
# ensembles, redução de dimensionalidade, etc.
```

```python
# medindo a acutácia (media dos acertos)

acc = clf.score(scaler.transform(X_test),y_test)
print('Acuracia',acc)
```

```python
# prevendo dados do conjunto de testes

pred = clf.predict(scaler.transform(X_test[:1]))
pred
```

```python
# verificando se a previsao foi correta

image = np.array(X_test.iloc[0].to_list(), dtype='float')
pixels = image.reshape((28, 28))
plt.imshow(pixels, cmap='gray')
plt.show()
```
